```python
#Find a string

def count_substring(string, sub_string):
    lenght=len(string)
    lenght_s=len(sub_string)
    counter=0
    for i in range(lenght-lenght_s+1):
        found_string=True
        for j in range(i,i+lenght_s):
            index=j-i
            if string[j]!=sub_string[index]:
                found_string=False
        if found_string==True:
            counter+=1
    return counter


if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

```python
#Text Wrap

import textwrap

def wrap(string, max_width):
    c=0
    result=""
    for character in string:
        if c/(max_width-1)==1:
            result+=character+"\n"
            c=0
        else:
            c+=1
            result+=character
        

    return result

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```

```python
#Alphabet Rangoli

def print_rangoli(size):

    alphabet = "abcdefghijklmnopqrstuvwxyz"
    line="-"
    order=list(range(size))
    order.reverse()
    for i in order:
        row=[line]*(size*2-1)
        for j in range(0,size-i):
            current_letter=alphabet[j+i]
            row[size-1-j]=current_letter
            row[size-1+j]=current_letter
        print(line.join(row))
        
    order=list(range(1,size))
    for i in order:
        
        row=[line]*(size*2-1)
        for j in range(0,size-i):
            current_letter=alphabet[j+i]
            row[size-1-j]=current_letter
            row[size-1+j]=current_letter
        print(line.join(row))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
#The Minion Game

def minion_game(string):
    vowels = 'AEIOU'
    lenght=len(string)
    Stuart=0
    Kevin=0
    for i in range(lenght):
        if string[i] in vowels:
            Kevin+=lenght-i
        else:
            Stuart+=lenght-i
    if Stuart>Kevin:
        print("Stuart",Stuart)
    if Stuart<Kevin:
        print("Kevin",Kevin)
    if Stuart==Kevin:
        print("Draw")
        
    

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
#Collections.deque()

from collections import deque
n = int(input())
d = deque()

for i in range(0, n):
    inp = input().split()
    command = inp[0]
    if len(inp) == 2:
        num = int(inp[1])
        if command == "append":
            d.append(num)
        elif command == "appendleft":
            d.appendleft(num)
    elif command == "pop":
        d.pop()
    elif command == "popleft":
        d.popleft()

print(*d,sep=" ")
```

```python
#Validating UID

def validate(uid):
    
    if len(uid)!=10:  
        return False
               
    used_char={}

    for char in uid:
        if char in used_char:
            used_char[char]+=1
        else:
            used_char[char]=1

    for number in used_char.values():
        if number>1: 
            return False
           
    digits="0123456789"
    upper_case="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    lower_case="abcdefghijklmnopqrstuvwxyz"
    n_digits=0
    n_upper=0

    for key in used_char.keys():
        if key in digits:
            n_digits+=1
        if key in upper_case:
            n_upper+=1
        if key not in digits+upper_case+lower_case:
            return False

    if n_digits<3 or n_upper<2:
        return False

    return True
   

for _ in range(int(input())):
    UID=input()
    if validate(UID):
        print("Valid")
    else:
        print("Invalid")   
        
```

```python
#Number Line Jumps
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the kangaroo function below.
#This is simple to solve if one deals with the mathematical problem first: a solution is found if and only if the following conditions are met 
def kangaroo(x1, v1, x2, v2):
    if (x2-x1)*(v2-v1) < 0:
        if (x2-x1)%(v2-v1) == 0:
            return ("YES")
    return("NO")
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    x1V1X2V2 = input().split()

    x1 = int(x1V1X2V2[0])

    v1 = int(x1V1X2V2[1])

    x2 = int(x1V1X2V2[2])

    v2 = int(x1V1X2V2[3])

    result = kangaroo(x1, v1, x2, v2)

    fptr.write(result + '\n')

    fptr.close()
```

```python
#Viral Advertising

#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the viralAdvertising function below.
def viralAdvertising(n):
    i=1
    n=n-1
    shared=5
    liked=math.floor(shared/2)
    cumulative=liked
    for i in range(n):
        shared=math.floor(shared/2)*3
        liked=math.floor(shared/2)
        cumulative+=liked
    return cumulative

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    result = viralAdvertising(n)

    fptr.write(str(result) + '\n')

    fptr.close()
```
